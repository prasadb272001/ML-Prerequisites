```python
#above plots using matplot
#operator description
import matplotlib.pyplot as plt
fig=plt.figure()
axis=fig.add_axes([0.5,0.1,0.8,0.8])
axis.plot(x,y,"r")
```

```python
#subgraphs
fig,axis=plt.subplots(nrows=1,ncols=2)
for ax in axis:
    ax.plot(x,y,'r')
```

```python
#Draw a pic or graph,inside another graph
#fig.add_axes([left,right,width,height of convas])
fig=plt.figure()
axis1=fig.add_axes([0.1,0.1,0.8,0.8])#big axes 0 to 1
axis2=fig.add_axes([0.2,0.5,0.4,0.3])#small convas 
axis1.plot(x,y,"r")
axis2.plot(y,x,"g")
```

```python
#adding colors to markers and incresing sizes and adding markers
fig,axis=plt.subplots(dpi=80)
axis.plot(x,x**2,'r',alpha=0.8,linewidth=1.5,linestyle='-',marker='o',markersize=5,markerfacecolor='blue')
axis.plot(x,x+3,'y',alpha=0.8,linewidth=2,linestyle='-.',marker='+',markersize=15,markerfacecolor='red')
axis.plot(x,x+5,'b',alpha=0.1,linewidth=2.5,linestyle=':',marker='*',markersize=20,markerfacecolor='black')
```

```python
#Set the convas grid and axis range
fig,Axes=plt.subplots(1,2,figsize=(10,5))
Axes[0].plot(x,x**2,x,x**3,lw=2)
Axes[0].grid(True)

Axes[1].plot(x,x**2,x,x**3,lw=2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])
Axes[1].grid(True)
```

```python
n=np.array([0,1,2,3,4,5])
fig,axes=plt.subplots(1,4,figsize=(16,5))

axes[0].set_title("scatter")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n,n**2,lw=2)

axes[2].set_title("bar")
axes[2].bar(n,n**2,align="center",alpha=0.5)

axes[3].set_title("fill_between")
axes[3].fill_between(x,x**2,x**3,color="green",alpha=0.5)
```

```python
#Draw counter image
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt
```

```python
delta=0.025
```

```python
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
x,y=np.meshgrid(x,y)
z1=np.exp(-x**2-y**2)
z2=np.exp(-(x-1)**2-(y-1)**2)
z=(z1-z2)*2
print(x)
print(y)
```

```python
fig,ax=plt.subplots()
cs=ax.contour(x,y,z)
ax.clabel(cs,inline=1,fontsize=10)
ax.set_title('counter map')
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig=plt.figure(figsize=(14,6))

#specify the 3D graphs
ax=fig.add_subplot(1,1,1,projection='3d')
ax.plot_surface(x, y, z, rstride=4,cstride=4,lw=0)
```
