# VECTOR OPERATIONS

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
```

```python
#Creation of vector
#plt.quiver(0,0,x axis size,y-axis size)
plt.quiver(0,0,8,8)
plt.show()
```

```python
plt.quiver(0,0,4,5, scale_units='xy',angles='xy',color='blue')
```

```python
#Add two vectors
plt.quiver(0,0,4,15, scale_units='xy',angles='xy',color='blue')
plt.quiver(0,0,-3,-16, scale_units='xy',angles='xy',color='red')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Addition of two vectors
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sum=vector_1+vector_2
print(sum)
```

```python
#IN GRAPH
plt.quiver(0,0,2,13, scale_units='xy',angles='xy',color='blue')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',color='red')
plt.quiver(0,0,5,1, scale_units='xy',angles='xy',color='green')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Substraction
vector_1=np.asarray([0,0,2,3])
vector_2=np.asarray([0,0,3,-2])
sub=vector_1-vector_2
print(sub)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',color='blue')
plt.quiver(0,0,3,-2, scale_units='xy',angles='xy',color='red')
plt.quiver(0,0,-1,5, scale_units='xy',angles='xy',color='green')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
#Multiplication
vector_1=np.asarray([0,0,2,3])
vector_2=2*vector_1
print(vector_2)
```

```python
plt.quiver(0,0,2,3, scale_units='xy',angles='xy',color='blue')
plt.quiver(0,0,4,6, scale_units='xy',angles='xy',color='red')
plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

```python
# Dot Product

a=np.array([2,3])
b=np.array([4,4])
a_dot_b=np.dot(a,b)
print(a_dot_b)
```

```python
a_cross_b=np.cross(a,b)
print(a_cross_b)
```

```python
c=a**2
d=b**2
sum=c+d
magnitude_of_b=np.sqrt(sum)
projection=(np.dot(a,b)/magnitude_of_b**2)
print("projection",projection)
```
